## Ejercicio 2: Clasificación y transformación de datos categóricos

En este ejercicio se clasifican las variables de un conjunto de datos según su tipo y se aplican técnicas de transformación para convertir variables categóricas en variables numéricas.

Se utilizan codificación ordinal para variables con jerarquía y One-Hot Encoding para variables sin orden, con el objetivo de preparar los datos para su uso en análisis y modelos de minería de datos.

```python
import pandas as pd

df2 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Nombre": ["Juan", "María", "Pedro", "Ana"],
    "Edad": [32, 28, 40, 35],
    "Categoria_Cliente": ["Oro", "Plata", "Platino", "Oro"],
    "Estado_Civil": ["Soltero", "Casado", "Viudo", "Casado"]
})

print(df2)
```

### Clasificación de tipos de datos

- **ID**: Numérico (identificador)
- **Nombre**: Categórico (texto)
- **Edad**: Numérico
- **Categoría Cliente**: Ordinal (tiene jerarquía: Plata < Oro < Platino)
- **Estado Civil**: Categórico (sin orden)

### Estrategia de transformación de datos

- **Categoría Cliente**: Se aplicará codificación ordinal, ya que existe un orden jerárquico (Plata < Oro < Platino).
- **Estado Civil**: Se aplicará One-Hot Encoding, ya que no existe un orden entre las categorías.

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

# DataFrame original limpio
df2 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Nombre": ["Juan", "María", "Pedro", "Ana"],
    "Edad": [32, 28, 40, 35],
    "Categoria_Cliente": ["Oro", "Plata", "Platino", "Oro"],
    "Estado_Civil": ["Soltero", "Casado", "Viudo", "Casado"]
})

# Codificación ordinal para Categoria_Cliente
orden_categoria = [["Plata", "Oro", "Platino"]]
encoder_ordinal = OrdinalEncoder(categories=orden_categoria)
df2["Categoria_Cod"] = encoder_ordinal.fit_transform(df2[["Categoria_Cliente"]])

# One-Hot Encoding para Estado_Civil
encoder_onehot = OneHotEncoder(sparse_output=False, drop="first")
encoded_estado = encoder_onehot.fit_transform(df2[["Estado_Civil"]])

df_encoded = pd.DataFrame(
    encoded_estado,
    columns=encoder_onehot.get_feature_names_out(["Estado_Civil"]),
    index=df2.index
)

# Unión resultados
df2_transformado = pd.concat([df2, df_encoded], axis=1)

print("Resultado de la transformación de datos:")
print()
print("La columna 'Categoria_Cod' representa la codificación ordinal de 'Categoria_Cliente'.")
print("Las columnas binarias de 'Estado_Civil' muestran la aplicación de One-Hot Encoding.")
print()
print(df2_transformado.to_string(index=False))
```

### Interpretación de la transformación

- La variable "Categoría Cliente" quedó representada numéricamente (0, 1, 2), respetando su jerarquía.
- La variable "Estado Civil" se descompuso en variables binarias, permitiendo su uso en modelos sin asumir orden.

### Conclusión

La transformación de variables categóricas permitió adaptar el dataset a un formato numérico adecuado para su análisis.

La correcta elección entre codificación ordinal y One-Hot Encoding fue clave para respetar la naturaleza de cada variable, evitando introducir interpretaciones incorrectas en los datos.

Este proceso es fundamental en minería de datos, ya que garantiza que los algoritmos puedan trabajar correctamente con la información disponible.
