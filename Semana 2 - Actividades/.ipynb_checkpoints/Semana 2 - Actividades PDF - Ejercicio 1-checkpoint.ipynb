## Ejercicio 1: Identificación, limpieza y validación de datos

En este ejercicio se analiza un conjunto de datos de clientes con el objetivo de identificar problemas de calidad, como valores faltantes, datos inválidos y posibles duplicados. 

A partir de este diagnóstico, se propone y aplica una estrategia de limpieza que permite corregir errores y preparar el dataset para su posterior uso en análisis o modelos de minería de datos.

```python
import pandas as pd

df = pd.DataFrame({
    "Cliente": [1, 2, 3, 4, 5],
    "Edad": [25, -40, 32, 22, 35],
    "Salario": [50000, 60000, 70000, None, 80000],
    "Estado_Civil": ["Soltero", "Casado", "Divorciado", "Soltero", "Viudo"],
    "Credito_Aprobado": ["Sí", "No", "Sí", "No", "Sí"]
})

print(df)
```

```python
# Valores faltantes
print("Faltantes:")
print(df.isnull().sum())

# Edades negativas (outliers evidentes)
print("\nEdades negativas:")
print(df[df["Edad"] < 0])
```

```python
duplicados = df[df.duplicated()]

if duplicados.empty:
    print("No se encontraron registros duplicados.")
else:
    print("Se encontraron registros duplicados:")
    print(duplicados)
```

### Problemas detectados en el dataset

Se identificaron los siguientes problemas en los datos:

- Existe un valor faltante en la columna **Salario**.
- Se detectó una edad negativa en la columna **Edad**, lo que representa un error de carga de datos.
- No se encontraron registros duplicados exactos.
- No se observan, en este ejemplo reducido, inconsistencias evidentes en las variables categóricas.

```python
# 1. Corregir edad negativa
df["Edad"] = df["Edad"].abs()

# 2. Completar valor faltante en Salario con la mediana
df["Salario"] = df["Salario"].fillna(df["Salario"].median())

print(df)
```

### Estrategia de limpieza aplicada

Para mejorar la calidad del dataset se aplicaron las siguientes acciones:

- La edad negativa se corrigió utilizando el valor absoluto, ya que representa un error de carga.
- El valor faltante en la columna **Salario** se completó con la mediana, porque es una medida robusta frente a valores extremos.
- Se verificó la existencia de registros duplicados, pero no se encontraron casos en este conjunto de datos.

### Conclusión

El análisis permitió detectar y corregir problemas básicos de calidad de datos, como valores faltantes y datos inválidos. 

La aplicación de técnicas simples de limpieza, como el uso del valor absoluto para corregir edades negativas y la imputación con la mediana en salarios, permitió obtener un dataset más consistente y confiable. 

Este proceso demuestra la importancia del preprocesamiento como etapa fundamental antes de cualquier análisis o modelado en minería de datos.
