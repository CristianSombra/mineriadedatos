## Ejercicio: Proceso KDD y CRISP-DM

### Ejercicio 1: Identificación de fases del proceso KDD

Se presentan las siguientes situaciones:

1. Un equipo descarga una base de datos de ventas de un supermercado.  
→ Corresponde a la fase de **Selección**, ya que se eligen los datos con los que se va a trabajar.

2. Se eliminan datos faltantes y se corrigen errores.  
→ Es **Preprocesamiento**, porque se están limpiando los datos para evitar problemas en el análisis.

3. Se aplica un algoritmo de clustering para segmentar clientes.  
→ Es **Minería de Datos**, ya que se están aplicando técnicas para descubrir patrones.

4. Se analizan los resultados obtenidos.  
→ Corresponde a **Evaluación**, porque se interpretan los resultados para ver si son útiles.

---

### Ejercicio 2: Aplicación del modelo CRISP-DM

Un banco quiere detectar fraudes y realiza las siguientes acciones:

1. Se analizan fraudes anteriores y se define el objetivo.  
→ **Comprensión del negocio**, porque se busca entender el problema y qué se quiere lograr.

2. Se limpian y transforman los datos.  
→ **Preparación de los datos**, ya que se dejan listos para el análisis.

3. Se prueban modelos de Machine Learning.  
→ **Modelado**, porque se aplican técnicas para encontrar patrones.

4. Se evalúa cuál modelo funciona mejor.  
→ **Evaluación**, donde se analiza el rendimiento de los modelos.

5. Se implementa un sistema automático.  
→ **Despliegue**, ya que el modelo se pone en uso real.

Posibles soluciones:

Transformación:

- Se podría normalizar la edad o las notas para compararlas mejor.  
- También convertir los datos a formato numérico limpio antes de analizarlos.

### Ejercicio 2: Creación de gráficos en Python

Datos:

| Producto | Precio | Cantidad_Vendida |
|----------|--------|------------------|
| A        | 10     | 50               |
| B        | 20     | 30               |
| C        | 15     | 40               |
| D        | 25     | 20               |

```python
import pandas as pd
import matplotlib.pyplot as plt

datos = {
    "Producto": ["A", "B", "C", "D"],
    "Precio": [10, 20, 15, 25],
    "Cantidad_Vendida": [50, 30, 40, 20]
}

df = pd.DataFrame(datos)

# Histograma de precios
plt.hist(df["Precio"])
plt.title("Distribución de precios")
plt.show()

# Gráfico de dispersión
plt.scatter(df["Precio"], df["Cantidad_Vendida"])
plt.xlabel("Precio")
plt.ylabel("Cantidad Vendida")
plt.title("Precio vs Cantidad Vendida")
plt.show()
```

## Ejercicio Integrador: Análisis de clientes (Telecomunicaciones)

### 1. Aplicación de KDD y CRISP-DM

Proceso KDD:

- Selección: tomar los datos de clientes (edad, plan, consumo, reclamos, estado).  
- Preprocesamiento: verificar errores, valores faltantes o inconsistentes.  
- Transformación: convertir variables categóricas como "Plan" y "Estado_Cuenta" a valores numéricos si es necesario.  
- Minería de datos: buscar patrones, por ejemplo clientes morosos o con alto consumo.  
- Evaluación: analizar si los patrones encontrados sirven para tomar decisiones.

CRISP-DM:

- Comprensión del negocio: detectar clientes con riesgo de mora o baja.  
- Comprensión de los datos: revisar qué información tenemos disponible.  
- Preparación: limpiar y transformar los datos.  
- Modelado: aplicar técnicas para encontrar patrones.  
- Evaluación: validar si los resultados son útiles.  
- Despliegue: usar los resultados para tomar decisiones (ej: retener clientes).

---

### 2. Preprocesamiento de datos

Posibles problemas:

- No hay valores faltantes en el ejemplo, pero podría haberlos en datos reales.  
- Variables categóricas (Plan, Estado_Cuenta) necesitan transformarse.  

Tratamiento:

- Codificar variables categóricas (ej: Básico=0, Premium=1, etc).  
- Verificar que no haya datos fuera de rango.

```python
import pandas as pd

datos = {
    "Cliente": [1, 2, 3, 4, 5],
    "Edad": [25, 40, 32, 22, 35],
    "Plan": ["Básico", "Premium", "Estándar", "Básico", "Premium"],
    "Consumo_GB": [5, 50, 10, 7, 45],
    "Reclamos": [1, 0, 2, 3, 0],
    "Estado_Cuenta": ["Pagado", "Pagado", "Moroso", "Moroso", "Pagado"]
}

df = pd.DataFrame(datos)
df
```

```python
df.isnull().sum()
```

```python
int(df.duplicated().sum())
```

Al revisar el conjunto de datos no se observan valores faltantes ni registros duplicados. 
Esto permite trabajar el análisis de forma directa, aunque en un caso real igualmente sería necesario revisar inconsistencias o valores fuera de rango.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histograma consumo
sns.histplot(df["Consumo_GB"], bins=5, kde=True)
plt.show()

# Estado de cuenta
sns.countplot(x=df["Estado_Cuenta"])
plt.show()
```

## Interpretación de resultados
Se observa que los clientes con mayor consumo suelen estar al día en los pagos.
Los clientes morosos tienen menor consumo y más reclamos.
Puede haber relación entre cantidad de reclamos y estado de cuenta.

Conclusión: Estos patrones pueden servir para identificar clientes en riesgo y tomar medidas como promociones o seguimiento más cercano.
